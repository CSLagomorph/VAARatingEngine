```python
query = ('select s.StudyPlanID,\
        s.GeneratedPlanID,\
        s.YearID,\
        s.QuarterID,\
        s.CourseID,\
        RTRIM(c.CourseNumber) as CourseNumber,\
        g.ParameterSetID,\
        p.EnrollmentTypeID,\
        p.JobTypeID,\
        p.MajorID,\
        m.Name as MajorName,\
        p.MaxNumberOfQuarters,\
        p.NumberCoreCoursesPerQuarter,\
        p.QuarterPreferenceID,\
        p.SchoolID,\
        sc.Name as SchoolName\
        from VsaDev.dbo.StudyPlan as s\
        join VsaDev.dbo.GeneratedPlan as g on s.GeneratedPlanID = g.GeneratedPlanID\
        join VsaDev.dbo.ParameterSet as p on g.ParameterSetID = p.ParameterSetID\
        join VsaDev.dbo.Major as m on m.MajorID = p.MajorID\
        join VsaDev.dbo.School as sc on sc.SchoolID = p.SchoolID\
        join VsaDev.dbo.Course as c on s.CourseID = c.CourseID;')

```

```python
classes = pd.read_sql(query, conn)
# classes = classes.drop('StudyPlanID',1)
# classes = classes.drop('ParameterSetID',1)
# classes = classes.drop('EnrollmentTypeID',1)
# classes = classes.drop('YearID',1)
# classes = classes.drop('QuarterID',1)
# classes = classes.drop('JobTypeID',1)
# classes = classes.drop('MaxNumberOfQuarters',1)
# classes = classes.drop('NumberCoreCoursesPerQuarter',1)
# classes = classes.drop('QuarterPreferenceID',1)


```

```python
classes["CourseNumber"] = classes["CourseNumber"].str.strip()
classes["CourseNumber"] = classes["CourseNumber"].map(lambda x: x.lstrip().rstrip())
classes["CourseNumber"] = classes["CourseNumber"].map(lambda x : x.split(" ")[0]+x.split(" ")[1])
```

```python
# classes.set_index('GeneratedPlanID').to_dict()
plans = {plan : g["CourseNumber"].tolist() for plan,g in classes.groupby("GeneratedPlanID")}
quarter = {plan : g["QuarterID"].tolist() for plan,g in classes.groupby("GeneratedPlanID")}
```

```python
majorsPref = {coll : (g["MajorName"].drop_duplicates().values.tolist()) for coll,g in classes.groupby("GeneratedPlanID")}
SchoolPref = {coll : (g["SchoolName"].drop_duplicates().values.tolist()) for coll,g in classes.groupby("GeneratedPlanID")}
planID = {coll : (g["GeneratedPlanID"].drop_duplicates().values.tolist()) for coll,g in classes.groupby("GeneratedPlanID")}


```

```python
from collections import defaultdict
 
 
dd = defaultdict(list)
 
for d in (planID,plans,quarter, majorsPref,SchoolPref): # you can list as many input dicts as you want here
    for key, value in d.items():
        dd[key].append(value)
 
print(dd)
```
